```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    message : str

def greeting_node(state: AgentState) -> AgentState:
    """Simple node that adds a gretting message to the state"""

    state['message'] = "Hey " + state['message'] + ", how is your day going?"
    return state

def compliment_node(state: AgentState) -> AgentState:
    """Simple node that adds a compliement node to the state"""

    state['message'] += " Answer: It's going very well!"
    return state
```

```python
graph = StateGraph(AgentState)

# Add a node to the graph

graph.add_node("greeter", greeting_node)
graph.add_node("compliment_node", compliment_node)

graph.set_entry_point("greeter")
graph.add_edge("greeter", "compliment_node")
graph.set_finish_point("compliment_node")

app = graph.compile()
```

```python
result = app.invoke({"message":"Bob"})
```
